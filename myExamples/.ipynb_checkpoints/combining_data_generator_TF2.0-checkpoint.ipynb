```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
print(tf.__version__)
%load_ext autoreload
%autoreload 2
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Convolution2D,MaxPooling2D,Dropout,Flatten
from my_classes import DataGenerator
```

```python
train_df=pd.read_csv('train.txt')
val_df=pd.read_csv('val.txt')
```

```python
partition={}
partition['train']=list(train_df['name'])
partition['valid']=list(val_df['name'])
```

```python
ids=list(train_df['name'])+list(val_df['name'])
t_labels=list(train_df['label'])+list(val_df['label'])
```

```python
labels={}
for id,lab in zip(ids,t_labels):
    labels[id]=lab
```

```python
# Parameters
params = {'dim': (224,224),
          'batch_size': 32,
          'n_classes': 2,
          'n_channels': 3,
          'shuffle': True}
training_generator = DataGenerator(partition['train'], labels, **params)
validation_generator = DataGenerator(partition['valid'], labels, **params)
```

```python
partition['valid']
```

```python
a,b=next(iter(training_generator))
```

```python
plt.imshow(a[0]);
plt.title(np.argmax(b[0]));
```

```python
def create_model():
    model = Sequential()
    model.add(Convolution2D(filters=16,kernel_size=(3,3),padding='same',input_shape=(224,224,3),
                           activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Convolution2D(filters=32,kernel_size=(3,3),padding='same',activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Flatten())
    model.add(Dense(units=100,activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(units=2,activation='softmax'))
    return model
```

```python
model=create_model()
model.summary()
```

```python
model(np.array(a[-2:],dtype=np.float32))
```

```python
def loss(y,yhat):
    return tf.keras.losses.categorical_crossentropy(y,yhat)

def grad(model,x,y):
    with tf.GradientTape() as tape:
        yhat=model(x)
        loss_value = loss(y,yhat)
    return tape.gradient(loss_value,model.trainable_variables)

def accuracy(y,yhat):
    
    yhat = tf.argmax(yhat,1).numpy()
    y=np.argmax(y,1)
    
    return np.sum(y==yhat)/len(y)
```

```python
optimizer = tf.keras.optimizers.Adam()
loss_history=tf.keras.metrics.Mean("loss")
accuracy_history = tf.keras.metrics.Mean("accuracy")
epochs=4
```

```python
for epoch in range(epochs):
    for x,y in training_generator:
        x=np.array(x,dtype=np.float32)
        yhat = model(x)
        grads = grad(model,x,y)
        optimizer.apply_gradients(zip(grads,model.trainable_variables))
        loss_history(loss(y,yhat))
        accuracy_history(accuracy(y,yhat))
        
    print("Epoch {:.3f} Loss {:.3f} Accuracy {:.3f}".format(epoch+1,loss_history.result(),
                                                accuracy_history.result()))
    loss_history.reset_states()
    accuracy_history.reset_states()
```

```python
for x,y in validation_generator:
    x=np.array(x,dtype=np.float32)
    yhat = model(x)
    accuracy_history(accuracy(y,yhat))
print("Accuracy of test set is {:.3f}".format(accuracy_history.result()))
```

```python
model.compile(optimizer=optimizer,loss=keras.losses.categorical_crossentropy,
             metrics=['accuracy'])
```

```python
model.fit_generator(generator=training_generator,
                    validation_data=training_generator,epochs=4)
```
